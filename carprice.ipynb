```python
OwnerMap = {
    'First Owner':1,
    'Second Owner':2,
    'Third Owner':3,
    'Fourth & Above Owner':4,
    'Test Drive Car':5
}

df['owner'] = df['owner'].map(OwnerMap)
```

```python
dropdata=df[df['fuel'].isin(['CNG', 'LPG'])].index
df=df.drop(dropdata)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["name"] = le.fit_transform(df["name"])
df["fuel"] = le.fit_transform(df["fuel"])
df["seller"] = le.fit_transform(df["seller"])
df["transmission"] = le.fit_transform(df["transmission"])
df["owner"] = le.fit_transform(df["owner"])
df["engine"] = le.fit_transform(df["engine"])
df["power"] = le.fit_transform(df["power"])
df["mileage"] = le.fit_transform(df["mileage"])

df["name"].unique()
df["fuel"].unique()
df["seller"].unique()
df["transmission"].unique()
df["owner"].unique()
df["engine"].unique()
df["power"].unique()
df["mileage"].unique()
```
